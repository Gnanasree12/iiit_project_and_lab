```python
# prompt: plot the cumulative variace of pca for all the possibel pronviopal components
## to do
import matplotlib.pyplot as plt
import numpy as np
pca = PCA()
pca.fit(numeric_data)
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

```python
pca = PCA(n_components=2)
pca.fit(numeric_data)
pca_data = pca.transform(numeric_data)

# Create a DataFrame with the principal components and target labels
pca_df = pd.DataFrame({
    "pca_1": pca_data[:, 0],
    "pca_2": pca_data[:, 1],
    "target": data["target"]
})

# Visualize the PCA results with a scatter plot
sns.scatterplot(x="pca_1", y="pca_2", hue="target", data=pca_df)
plt.title("PCA Visualization of Heart Disease Data")
plt.show()
```

```python
from sklearn.manifold import TSNE

# Initialize and fit the TSNE model
tsne = TSNE(n_components=2)
tsne_data = tsne.fit_transform(numeric_data)

# Create a DataFrame with the TSNE components and target labels
tsne_df = pd.DataFrame({
    "tsne_1": tsne_data[:, 0],
    "tsne_2": tsne_data[:, 1],
    "target": data["target"]
})

# Visualize the TSNE results with a scatter plot
sns.scatterplot(x="tsne_1", y="tsne_2", hue="target", data=tsne_df)
plt.title("TSNE Visualization of Heart Disease Data")
plt.show()
```

```python
##  one hot encoding of categorical features in data
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
```

```python
data = onehot_encode(
    data,
    columns=['Beverage', 'Beverage_prep'],
    prefixes=['bev', 'bevp']
)
```

```python
## to do

# Create a pie chart of the 'Tea' column also write your observation form the plot
import pandas as pd
import altair as alt

# Create a DataFrame with the data provided in the query with the columns `Tea` and `Frequency`.
data = {'Tea': ['Green', 'Black', 'Herbal', 'Oolong', 'Other'],
        'Frequency': [50, 35, 25, 10, 5]}
df = pd.DataFrame(data)

# Calculate the percentage of each value in relation to the total sum of the `Frequency` column, store it as a number between 0 and 1.
df['percentage'] = df['Frequency'] / df['Frequency'].sum()
base = alt.Chart(df).encode(theta=alt.Theta('Frequency:Q', stack=True)).properties(title='Tea Frequency')

# Create a pie chart by specifying the type of mark as an arc.
# Specify the outer radius of the arcs and encode color based on the `Tea` column.
pie = base.mark_arc(outerRadius=120).encode(
    color=alt.Color('Tea:N'),
    # Order the arcs by `percentage` in descending order.
    order=alt.Order('percentage', sort='descending'),
    # Add tooltips for additional information.
    tooltip=['Tea', 'Frequency', alt.Tooltip('percentage', format='.1%')],
)

text = base.mark_text(radius=140).encode(
    text=alt.Text('percentage', format='.1%'),
    # Order of the text should match the order of the arcs.
    order=alt.Order('percentage', sort='descending'),
    color=alt.value('black')  # Set the color of the labels to black
)

chart = pie + text

# Save the chart in a JSON file
chart.save('tea_frequency_pie_chart.json')
```
